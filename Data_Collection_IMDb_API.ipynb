## The purpose of writing this API calls to IMDb databse was to retrive crucial information like budhget and box office of all the movies in my bollywood_full.csv dataset

## Getting busget and box offcie attributes for the dataset would ahve revealed a lot of important factors like correlation bettwen runtimes, movies' time of release, relationsio bwtween cast/crew and box office etc.



```python
import pandas as pd
import numpy as np
# from lib import imdb
from imdb import IMDb
import re
ia = IMDb()


def read_data():
    '''
    This function reads a csv file containing imdb_ids and returns it in a dataframe.
    '''
    df = pd.read_csv('/content/bollywood_crew.csv')
    to_drop = ['directors', 'writers']
    df.drop(to_drop, axis = 1, inplace = True)
    df.reset_index()
    return df

def post_api_call(imdb_id):
    '''
    This function calls the IMDB API and fetches values for director, cast, budget, box office and reviews and returns these values.
    '''
    try:
        movie_dict = ia.get_movie(imdb_id)
    except Exception:
        return np.nan, np.nan, np.nan, np.nan, False
    try:
        box_office = movie_dict['box office']['Cumulative Worldwide Gross']
        box_office = re.sub(r"[^0-9]+", '', box_office)
    except Exception:
        box_office = np.nan
    try:
        budget = movie_dict['box office']['Budget']
        budget = re.sub(r"[^0-9]+", '', budget)
    except Exception:
        budget = np.nan
    
    return box_office, budget

def create_new_dataset(df):
    '''
    This function adds columns to the dataframe to accomodate director, cast, budget and box office. It also creates a separate csv file to store reviews.
    '''
    df['box_office'] = ['' for x in range(len(df))]
    df['budget'] = ['' for x in range(len(df))]

    for i in range(len(df)):
        print("Processing movie : {}".format(i + 1))
        # box_office, budget = post_api_call(df.at[i, 'imdb_id'])
        box_office, budget = post_api_call(df.at[i, 'imdb_id'])
        # budget = post_api_call(df.at[i, 'imdb_id'])

        df.at[i, 'budget'] = budget
        df.at[i, 'box_office'] = box_office

    df.to_csv("data_with_revenue.csv", index = False) 

def main():
    df = read_data()
    create_new_dataset(df)

if __name__ == "__main__":
    main()
```
